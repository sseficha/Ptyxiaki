```python
from datetime import datetime
df = pd.read_csv('../datasets/headline_embeddings.csv', index_col='date', parse_dates=['date'])
df = df.loc[(df.index >= datetime(2014, 1, 1, 00, 00, 00))]

print(df)
```

```python
df = df.resample('D').mean()
df = df.ffill().bfill()
print(df)
# df.to_csv('../datasets/headline_embeddings_mean.csv')
```

```python
#scale
scaler = MinMaxScaler((-1,1))
for i in range(len(df.columns)):
    df.loc[:, str(i)] =scaler.fit_transform(df.loc[:,str(i)].values.reshape(-1,1))

print(df)
# df.to_csv('../datasets/headline_embeddings_mean.csv')
```
