```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./sentiment.csv', parse_dates=['datetime'], header=0)
plt.figure(figsize=(20,10))
sb.lineplot(x=df['datetime'], y=df['score'])
```
